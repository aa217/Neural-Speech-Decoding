```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Read the CSV file from the data folder
data_path = os.path.join('data', 'food_1.csv')
df = pd.read_csv(data_path)

# Get the first column
first_column = df.iloc[:, 0]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(first_column)
plt.title('EEG Data - First Column')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Show the plot
plt.show()

# Print some basic info about the data
print(f"Data shape: {df.shape}")
print(f"First column name: {df.columns[0]}")
print(f"First few values: {first_column.head()}")
```

```python
# Read the CSV file from the data folder
data_path = os.path.join('data', 'food_2.csv')
df = pd.read_csv(data_path)

# Get the first column
first_column = df.iloc[:, 0]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(first_column)
plt.title('EEG Data - First Column')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Show the plot
plt.show()

# Print some basic info about the data
print(f"Data shape: {df.shape}")
print(f"First column name: {df.columns[0]}")
print(f"First few values: {first_column.head()}")
```

```python
# Get the transposed dataframe
df_transposed = df.transpose()

# Plot the first row of the transposed data
plt.figure(figsize=(12, 6))
plt.plot(df_transposed.iloc[0, :])
plt.title('Transposed Data - First Row')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Transposed data shape: {df_transposed.shape}")
print(f"First row values (first 10): {df_transposed.iloc[0, :10].values}")
```

```python
import mindsai_filter_python as mai

tailoring_lambda = 1e-29
filtered_data = mai.mindsai_python_filter(df_transposed.values, tailoring_lambda)

# Turn Numpy array back to dataframe
filtered_data = pd.DataFrame(filtered_data, columns=df_transposed.columns)
```

```python
scaled_filtered_data = filtered_data
```

```python
plt.figure(figsize=(12, 6))
plt.plot(scaled_filtered_data.iloc[0, :])
plt.plot(df_transposed.iloc[0, :])
#legend
plt.legend(['Filtered Data', 'Original Data'])
plt.title('Food_1 - Filtered vs Original Data (First channel)')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
